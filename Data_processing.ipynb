```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Caithy'],
    'Age': [19, 18, 22],
    'Gender': ['F', 'M', 'F'],
    'Marks': [85, 65, 85]
}

df = pd.DataFrame(data)
print(df)
```

```python
def grades(marks):
     if marks >= 85:
        return 'A'
     elif marks >= 70:
        return 'B'
     else:
        return 'C'
df['Grades']=df['Marks'].apply(grades)
print(df)
```

```python
bins = [0, 18, 21, 100]
labels = ['Teen', 'Young Adult', 'Adult']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
print(df)
```

```python
data=pd.read_csv('Name-Age-Gender-Marks-City - Sheet1.csv')
data = data[(data['Marks']>=60)&(data['Marks']<=100)]
print(data)
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data=train_test_split(data,test_size=0.2,random_state=42)
print("Training Data:")
print(train_data)
print("\nTesting Data:")
print(test_data)
```
